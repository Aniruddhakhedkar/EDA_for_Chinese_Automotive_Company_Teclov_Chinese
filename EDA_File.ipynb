### Handling the outliers-(Here outliers are handles w.r.p. continuous variabels and hence x & y are the required dataframes)

```python
x.skew()
```

```python
#Outlier handling for Engine_Size-

sns.boxplot(data=x['Engine_Size'])
```

```python
#By using zscore method-
zs=stats.zscore(x['Engine_Size'])
z_abs=np.abs(zs)
x['Engine_Size_C']=x['Engine_Size'][z_abs<3]
x['Engine_Size_C1']=x['Engine_Size_C'].fillna(x['Engine_Size_C'].median())
x['Engine_Size_C1'].isnull().sum()
x['Engine_Size_C1'].skew()
```

```python
#By using capping method-
Q1=x['Engine_Size'].quantile(0.25)
Q3=x['Engine_Size'].quantile(0.75)
IQR=Q3-Q1

UL=Q3+(1.5*IQR)
LL=Q1-(1.5*IQR)

x['Engine_Size_C2']=np.where(x['Engine_Size']>UL,UL,x['Engine_Size'])
x['Cleaned_Engine_Size']=np.where(x['Engine_Size_C2']<LL,LL,x['Engine_Size_C2'])

sns.boxplot(x['Cleaned_Engine_Size'])
x['Cleaned_Engine_Size'].skew()
```

```python
#Outlier handling for Horse_Power-

#By using zscore method-
sns.boxplot(x['Horse_Power'])
zs=stats.zscore(x['Horse_Power'])
z_abs=np.abs(zs)
x['Horse_Power_C']=x['Horse_Power'][z_abs<3]
x['Horse_Power_C'].info()
```

```python
#By using capping method-
Q1=x['Horse_Power'].quantile(0.25)
Q3=x['Horse_Power'].quantile(0.75)
IQR=Q3-Q1

UL=Q3+(1.5*IQR)
LL=Q1-(1.5*IQR)

x['Horse_Power_C2']=np.where(x['Horse_Power']>UL,UL,x['Horse_Power'])
x['Cleaned_Horse_Power']=np.where(x['Horse_Power_C2']<LL,LL,x['Horse_Power_C2'])

sns.boxplot(x['Cleaned_Horse_Power'])
x['Cleaned_Horse_Power'].skew()
```

```python
x_cleaned=x.drop(['Engine_Size','Horse_Power','Engine_Size_C','Engine_Size_C2','Horse_Power_C','Engine_Size_C1','Horse_Power_C2'],axis=1)
x_cleaned.skew()
```

In the y dataframe, Compression_Ratio and Price showcases high skeness (>1.0), hence the presence of significant outliers-

```python
sns.boxplot(y['Compression_Ratio'])
```

```python
#Removal of outlier for Compression_Ratio by using capping method-
Q1=y['Compression_Ratio'].quantile(0.25)
Q3=y['Compression_Ratio'].quantile(0.75)
IQR=Q3-Q1

UL=Q3+(1.5*IQR)
LL=Q1-(1.5*IQR)

y['Compression_Ratio_1']=np.where(y['Compression_Ratio']>UL,UL,y['Compression_Ratio'])
y['Cleaned_Compression_Ratio']=np.where(y['Compression_Ratio_1']<LL,LL,y['Compression_Ratio_1'])

sns.boxplot(y['Cleaned_Compression_Ratio'])
y['Cleaned_Compression_Ratio'].skew()
```

```python
#Removal of outlier for Price by using capping method-
Q1=y['Price'].quantile(0.25)
Q3=y['Price'].quantile(0.75)
IQR=Q3-Q1

UL=Q3+(1.5*IQR)
LL=Q1-(1.5*IQR)

y['Price_1']=np.where(y['Price']>UL,UL,y['Price'])
y['Cleaned_Price']=np.where(y['Price_1']<LL,LL,y['Price_1'])

sns.boxplot(y['Cleaned_Price'])
y['Cleaned_Price'].skew()
```
